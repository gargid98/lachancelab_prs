```python
# This is the prs script used for 225 southeastern european ancient samples using breast cancer risk alleles
```

```python
#read in risk allele file
riskallelefile = open("PGS000004.txt", "r")
print("reading in risk allele file")
riskallele=[]
for line in riskallelefile:
    if line.startswith("#"):
        continue
    if line.startswith("chr"):
        continue
    else:
        row = line.strip().split("\t")
        riskallele.append(row)

```

```python
#reading in rsid file pulled from kaviar: to convert 
print("reading in rsid file")
rsidfile = open("mavaddatrsids.txt", "r")
rsid=[]
for line in rsidfile:
    if line.startswith("#"):
        continue
    if line.startswith("c"):
        continue
    else:
        row = line.strip().split("\t")
        rsid.append(row)
```

```python
print("creating rsid list")

#joining kaviar snps files with risk allele file, replacing chrom pos with rsid
rsidlist_all=[]       
for i in riskallele: 
    for j in rsid:
        if [i[0], i[1]] ==  [j[0], j[1]]:
            if [i[3], i[2]] == [j[3], j[4]]:
                rsidlist_all.append([j[2], i[2], i[3], i[4]])
                
#removing rsid duplicates
rsidlist=[]       
rsids=[]    
for i in rsidlist_all:   
    if i[0] in rsids:
        continue
    else:
        rsidlist.append(i)
    rsids.append(i[0])
```

```python
#reading in the vcf file 
file1open = open("new225.vcf", "r")
print("reading in vcf file")
vcffile=[]
individuals=[]
for line in file1open:
    if line.startswith("##"):
        continue
    if line.startswith("#CHR"): 
        inds = line.strip().split("\t")
        individuals.append(inds)
        vcffile.append(inds)
    row = line.strip().split("\t")
    if row[2] in rsids:
        if row[2] not in vcffile:
            vcffile.append(row)
print("number of risk alleles found in file", len(vcffile))
```

```python
vcf=[]
unique=[]
for i in vcffile:
    if i[2] in unique:
        continue
    else:
        vcf.append(i)
        unique.append(i[2])
print("number of unique rsids", len(vcf))
```

```python
print("calculating prs")

scores=[]
null=0
genotype=0
#iterate through individuals
for j in range(9, len(inds)):
        sample=[]
        totalscore=0
        #iterate through risk alleles
        for i in range(1, len(vcf)):
            #sample contains rsid, reference allele, alternate allele, and genotype
            sample.append([vcf[i][2], vcf[i][3], vcf[i][4], vcf[i][j]])
        for i in sample:
            for x in rsidlist:
                if i[0] == x[0]:
                    #checking if ref allele and alt allele match
                    if i[1]==x[2] and i[2]==x[1]:
                        if i[3]!="./.": 
                            if i[3] == "0/0": 
                                genotype = 0
                            if i[3] == "0/1" or "1/0": 
                                genotype = 1 
                            if i[3] == "1/1": 
                                genotype = 2 
                            score=float(x[3])* float(genotype)
                            totalscore=score + totalscore
                        else:
                            null+=1
                    #if ref allele and alternate allele are reversed
                    elif i[1]==x[1] and i[2]==x[2]:
                        if i[3]!="./.": 
                            if i[3] == "0/0": 
                                genotype = 2
                            if i[3] == "0/1" or "1/0": 
                                genotype = 1 
                            if i[3] == "1/1": 
                                genotype = 0 
                            score=float(x[3])* float(genotype)
                            totalscore=score + totalscore
                        else:
                            null+=1
                    else:
                        print("null")
        scores.append([individuals[0][j], totalscore])
        
```

```python
for i in scores: 
    print(i[0] + ": " + str(i[1]))
```
